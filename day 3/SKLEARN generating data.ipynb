```python
from sklearn.datasets import make_regression, make_classification, make_blobs, make_moons
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x , y = make_regression?
```

```python
x , y = make_regression(n_samples=400, n_features=1, n_informative=1 , noise=10)
```

```python
x.shape , y.shape
```

```python
plt.scatter(x, y , color="blue")
plt.show()
```

```python
xc , yc = make_classification(n_samples=400, 
                              n_features=2, 
                              n_classes=4, 
                              n_clusters_per_class=1,
                              n_informative=2, 
                              n_redundant=0)
```

```python
print(xc.shape , yc.shape)
print(np.unique(yc))
yc[:10]
```

```python
plt.scatter(xc[:,0] , xc[:,1] , c = yc)
```

```python
xb , yb = make_blobs(n_samples=400 , n_features=2 , centers=5 )
```

```python
print(xb.shape , yb.shape)
```

```python
plt.scatter(xb[:,0] ,  xb[:,1] , c= yb)
```

```python
xm , ym = make_moons(n_samples=400 , shuffle=True, noise=0.2, random_state=1 )
```

```python
print(xm.shape , ym.shape)
```

```python
plt.scatter(xm[:,0] ,  xm[:,1] , c= ym)
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2 , 2 , 1)
plt.scatter(x, y , color="yellow")

plt.subplot(2 , 2, 2)
plt.scatter(xc[:,0] , xc[:,1] , c = yc)

plt.subplot(2 , 2 , 3)
plt.scatter(xb[:,0] ,  xb[:,1] , c= yb)

plt.subplot(2 ,2 ,4)
plt.scatter(xm[:,0] ,  xm[:,1] , c= ym)

plt.show()
```
