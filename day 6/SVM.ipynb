```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
X_train = pd.read_csv("train.csv")
```

```
X_train.head()
```

```
X_train.columns
```

```
l1 = ['PassengerId', "Name", "Ticket"]
X_train.drop(l1, axis = 1, inplace = True)
```

```
X_train.head()
```

```
X_train.info()
```

```
l2 = ["Cabin"]
X_train.drop(l2, axis=1, inplace = True)
X_train.info()
```

```
X_train.isnull().sum()
```

```
emb_fill = list(X_train["Embarked"].mode())[0]
print(emb_fill)
print(type(emb_fill))
X_train["Age"].fillna(X_train["Age"].mean(), inplace = True)
X_train["Embarked"].fillna(emb_fill, inplace = True)
```

```
X_train.isnull().sum()
```

```
X_train.shape
```

```
X_train.head()
```

```
from sklearn.preprocessing import LabelEncoder
le_g = LabelEncoder()
le_e = LabelEncoder()
```

```
X_train["Sex"] = le_g.fit_transform(X_train["Sex"])
X_train["Embarked"] = le_e.fit_transform(X_train["Embarked"])
```

```
X_train.head()
```

```
X = X_train.iloc[:, 1:].values
Y = X_train.iloc[:, 0].values
X.shape, Y.shape
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```
X[:5]
```

```
from sklearn.linear_model import LogisticRegression
```

```
clf = LogisticRegression(solver= 'sag' ,n_jobs= -1)
```

```
clf.fit(X, Y)
```

```
Y_pred = clf.predict(X)
```

```
clf.score(X, Y)
```

```
from sklearn.metrics import confusion_matrix
```

```
print(confusion_matrix(Y_pred, Y))
```

```
pred_y1 = clf1.predict(X)
```
